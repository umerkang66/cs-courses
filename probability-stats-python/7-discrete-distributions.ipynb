```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

### Bernoulli Distributions

```python
from scipy.stats import bernoulli, binom, poisson
```

```python
data_bern = bernoulli.rvs(size=10000, p=0.6) # rsv: random variables
# p is success rate of a single variable

ax = sns.histplot(data_bern, bins=100, kde=False)
ax.set(xlabel="Bernoulli Distribution", ylabel="Frequency")

plt.show()
```

### Binomial Distributions

```python
data_binom = binom.rvs(n=10, size=1000, p=0.6) # rsv: random variables
# p is success rate of a single variable

ax = sns.histplot(data_binom, bins=100, kde=False)
ax.set(xlabel="Binomial Distribution", ylabel="Frequency")

plt.show()
```

### Poisson Distribution

```python
data_poisson = poisson.rvs(mu=3, size=10000) # rsv: random variables
# p is success rate of a single variable

ax = sns.histplot(data_poisson, bins=100, kde=False)
ax.set(xlabel="Poisson Distribution", ylabel="Frequency")

plt.show()
```
