```python
import numpy as np
import matplotlib.pyplot as plt
```

### Exercise 1: Implement the Function

$$f(x) = cos^2(x^2) + \pi$$

```python
def get_fx(x):
  return np.cos(x*x)**2 + np.pi
```

```python
x = np.linspace(-2.1, 2.1, 1000)
fx = get_fx(x)
```

```python
plt.plot(x, fx)

plt.xlabel("x")
plt.ylabel("$f(x) = cos^2(x^2) + \\pi$")
plt.grid()
plt.show()
```

### Computing the Left+Right Limit by Py.

$$\lim_{x \to a} f(x) = ?$$

$a = 1$

Start with $x_o = [a-1,a+1]$

```python
def lim_zeno_paradox(fx, a, iterations = 10):
  x0 = np.array([a-1,a+1])

  left_right_lims = np.zeros((iterations, 2))
  x_axis_vals = np.zeros((iterations, 2))

  for i in range(iterations):
    x_axis_vals[i,:] = x0
    left_right_lims[i,:] = fx(x0)
    # update the x0-values
    x0 = (x0+a)/2

  return left_right_lims, x_axis_vals
```

```python
iterations = 10
a = 1
lims, x_axis = lim_zeno_paradox(get_fx, a=a,iterations=iterations)
# first column is limit from the left, and second column is the limit from the right

# after many iterations, the left and right limits will be equal, thus means that the limit exist
print("Limit from the left:")
print(np.vstack((x_axis[:,0], lims[:,0])).T)

print("Limit from the right:")
print(np.vstack((x_axis[:,1], lims[:,1])).T)

print(f"Value of f(x) at x=1")
print(get_fx(a))
```

### Visualization the Results

```python
plt.plot(x,get_fx(x),'k')
plt.plot([a,a], [np.pi, 1+np.pi], 'k--')
# these have two columns, so matplotlib is plotting first column with different color and other column with different color
plt.plot(x_axis, lims, 'o', markerfacecolor='w')

plt.title(f"f({a}) = {get_fx(a)}")
plt.show()
```
