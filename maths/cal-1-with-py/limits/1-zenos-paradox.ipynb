### Exercise 1: Implement the Function

$$f(x) = cos^2(x^2) + \pi$$

```python
def get_fx(x):
  return np.cos(x*x)**2 + np.pi
```

```python
x = np.linspace(-2.1, 2.1, 1000)
fx = get_fx(x)
```

```python
plt.plot(x, fx)

plt.xlabel("x")
plt.ylabel("$f(x) = cos^2(x^2) + \\pi$")
plt.grid()
plt.show()
```

### Computing the Left+Right Limit by Py.

$$\lim_{x \to a} f(x) = ?$$

$a = 1$

Start with $x_o = [a-1,a+1]$

```python
def lim_zeno_paradox(fx, a, iterations = 10):
  x0 = np.array([a-1,a+1])

  left_right_lims = np.zeros((iterations, 2))

  for i in range(iterations):
    left_right_lims[i,:] = fx(x0)
    # update the x0-values
    x0 = (x0+a)/2

  return left_right_lims
```

```python
iterations = 10
lims = lim_zeno_paradox(get_fx, a=1,iterations=iterations)
# first column is limit from the left, and second column is the limit from the right

# after many iterations, the left and right limits will be equal, thus means that the limit exist
lims
```
