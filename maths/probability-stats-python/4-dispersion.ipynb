### Dispersion

Average *SQUARED distance* of values from Mean. The more the distance, the more the *Dispersion*.

Why can't we just take the Absolute value, because absolute value can't be differentiable.


<center>
<h4>MEAN</h4>
</center>

$$\overline{X} = (\sum_{i=1}^{N} x_i) / N$$


<center>
<h4>VARIANCE Squared</h4>
</center>

$$\sigma^2 = \frac{\sum_{i=1}^{N} (x_i - \overline{x})^2}{N}$$

where $\overline{x}$ is MEAN, $x_i$ is distance of point $i_{th}$ element $x$ from Mean.


<center>
<h4>Standard Deviation</h4>
</center>

$$\sigma = \sqrt{\sigma^2}$$

In Variance or Sigma Squared, the unit becomes squared, hence the solution is standard deviation.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
sns.set_style("white")
```

### Example Random Numbers

```python
np.random.uniform(low=0.0, high=1.0)
```

```python
np.random.uniform(low=0.0, high=1.0, size=10) # give me 10 values in an array
```

### A LOT of Random Numbers

```python
num_samples = 10000 # get rid of magic_numbers

uniform_vals  = np.random.uniform(low=0.0, high=10.0, size=num_samples)
```

```python
sns.histplot(uniform_vals, bins=20, kde=True)

plt.ylabel("Frequency")
plt.xlabel("Value")
plt.title("Uniform Values")

sns.despine(offset=10)

plt.show()
```

```python
normal_values = np.random.normal(loc=5.0, scale=15.0, size=num_samples) # meaning most values should be close to 5 and more you move away from 5, the more the frequency should decrease

# 15 is dispersion
```

```python
sns.histplot(normal_values, bins=20, kde=True)

plt.ylabel("Frequency")
plt.xlabel("Value")
plt.title("Normal Values")

sns.despine(offset=10)

plt.show()
```

```python
print("Uniform Vals Mean: ", np.mean(uniform_vals))
print("Normal Vals Mean: ", np.mean(normal_values))
```

```python
print("Uniform Vals Variance: ", np.var(uniform_vals))
print("Normal Vals Variance: ", np.var(normal_values)) # this has more spread, means the more variance
```

```python
print("Uniform Vals Standard Deviation: ", np.std(uniform_vals))
print("Normal Vals Standard Deviation: ", np.std(normal_values)) # same with the standard deviation
```

```python
sns.histplot(normal_values, bins=20)

plt.ylabel("Frequency")
plt.xlabel("Value")
plt.title("Normal Values")

mean = np.mean(normal_values)
var = np.var(normal_values)
std = np.std(normal_values)

# plotting the mean
plt.plot([mean, mean],[0, 1600], color='red', linewidth=2, label="Mean")

# plotting the variance
plt.plot([mean + std, mean + std],[0, 1600], color='green', linewidth=2, label="Mean + Std")

# plotting the standard variation
plt.plot([mean - std, mean - std],[0, 1600], color='blue', linewidth=2, label="Mean - Std")

sns.despine(offset=10)
plt.legend()
plt.xlim(-60, 60)
plt.show()
```

### Increasing the dispersion

```python
normal_values = np.random.normal(loc=5.0, scale=20.0, size=num_samples) # meaning most values should be close to 5 and more you move away from 5, the more the frequency should decrease
```

```python
sns.histplot(normal_values, bins=20)

plt.ylabel("Frequency")
plt.xlabel("Value")
plt.title("Normal Values")

mean = np.mean(normal_values)
var = np.var(normal_values)
std = np.std(normal_values)

# plotting the mean
plt.plot([mean, mean],[0, 1600], color='red', linewidth=2, label="Mean")

# plotting the variance
plt.plot([mean + std, mean + std],[0, 1600], color='green', linewidth=2, label="Mean + Std")

# plotting the standard variation
plt.plot([mean - std, mean - std],[0, 1600], color='blue', linewidth=2, label="Mean - Std")

sns.despine(offset=10)
plt.legend()
plt.xlim(-60, 60)
plt.show()
```

```python
# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

normal_values = np.random.normal(loc=5.0, scale=15.0, size=num_samples)

# Plot on the first subplot
sns.histplot(normal_values, bins=20, ax=axs[0])
axs[0].set_ylabel("Frequency")
axs[0].set_xlabel("Value")
axs[0].set_title("Less Variant")

# Plot the mean and standard deviations
axs[0].plot([mean, mean], [0, 1600], color='red', linewidth=2, label="Mean")
axs[0].plot([mean + std, mean + std], [0, 1600], color='green', linewidth=2, label="Mean + Std")
axs[0].plot([mean - std, mean - std], [0, 1600], color='blue', linewidth=2, label="Mean - Std")
axs[0].legend()
axs[0].set_xlim(-60, 60)
sns.despine(ax=axs[0], offset=10)




normal_values = np.random.normal(loc=5.0, scale=20.0, size=num_samples)

# Plot on the second subplot
sns.histplot(normal_values, bins=20, ax=axs[1], color='orange')
axs[1].set_ylabel("Frequency")
axs[1].set_xlabel("Value")
axs[1].set_title("More Variant")

# Plot the mean and standard deviations
axs[1].plot([mean, mean], [0, 1600], color='red', linewidth=2, label="Mean")
axs[1].plot([mean + std, mean + std], [0, 1600], color='green', linewidth=2, label="Mean + Std")
axs[1].plot([mean - std, mean - std], [0, 1600], color='blue', linewidth=2, label="Mean - Std")
axs[1].legend()
axs[1].set_xlim(-60, 60)
sns.despine(ax=axs[0], offset=10)

# Customize the second plot (optional)
axs[1].set_xlim(-60, 60)
sns.despine(ax=axs[1], offset=10)

# Show the plots
plt.tight_layout()
plt.show()
```

### Coefficient of Variation

$$\frac{\sigma}{\overline{x}} \times 100$$

Relative measure of deviation, percentage of Deviation with respect to Mean.
