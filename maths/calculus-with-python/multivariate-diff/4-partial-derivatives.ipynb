## Partial differentiation


```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

from IPython.display import display, Math
```

### Differentiate a multivariable function


```python
x, y = sym.symbols("x,y")

# create the function
fxy = x**2 / sym.pi**3 + sym.sin(y)
display(Math("f(x,y) = %s" % sym.latex(fxy)))
```

```python
# partial derivatives
df_x = sym.diff(fxy, x)
display(Math(r"\frac{\partial f}{\partial x} = %s" % sym.latex(df_x)))
print(" ")


df_y = sym.diff(fxy, y)
display(Math(r"\frac{\partial f}{\partial y} = %s" % sym.latex(df_y)))
```

### Visualize the function and its partial derivatives


```python
# lambdify expressions
fxy_lam = sym.lambdify((x, y), fxy)
dfx_lam = sym.lambdify((x, y), df_x)
dfy_lam = sym.lambdify((x, y), df_y)


# and evaluate at specific points
xx = np.linspace(0, 2 * np.pi, 21)
X, Y = np.meshgrid(xx, xx)
```

```python
# now draw them
_, axs = plt.subplots(1, 3, figsize=(8, 5))
axs[0].imshow(fxy_lam(X, Y), extent=[xx[0], xx[-1], xx[0], xx[-1]], origin="upper")
axs[0].set_title(f"$f(x,y) = {sym.latex(fxy)}$")


axs[1].imshow(dfx_lam(X, Y), extent=[xx[0], xx[-1], xx[0], xx[-1]], origin="upper")
axs[1].set_title(f"$f_x = {sym.latex(df_x)}$")


axs[2].imshow(dfy_lam(X, Y), extent=[xx[0], xx[-1], xx[0], xx[-1]], origin="upper")
axs[2].set_title(f"$f_y = {sym.latex(df_y)}$")

for a in axs:
    a.set_xlabel("x")
    a.set_ylabel("y")

plt.tight_layout()
plt.show()
```

### Now with a cross-term


```python
# Define symbols
x, y = sym.symbols("x y")

# Function
fxy = x**2 + sym.sin(y) + x**2 * sym.sin(y)
display(Math(r"f(x,y) = %s" % sym.latex(fxy)))


# Partial derivatives
df_x = sym.diff(fxy, x)
print(" ")
display(Math(r"\frac{\partial f}{\partial x} = %s" % sym.latex(df_x)))
print(" ")

df_y = sym.diff(fxy, y)
display(Math(r"\frac{\partial f}{\partial y} = %s" % sym.latex(df_y)))
```

```python
# lambdify expressions
fxy_lam = sym.lambdify((x, y), fxy)
dfx_lam = sym.lambdify((x, y), df_x)
dfy_lam = sym.lambdify((x, y), df_y)


# now draw them
_, axs = plt.subplots(1, 3, figsize=(8, 5))
axs[0].imshow(fxy_lam(X, Y), extent=[xx[0], xx[-1], xx[0], xx[-1]], origin="upper")
axs[0].set_title(f"$f(x,y) = {sym.latex(fxy)}$")


axs[1].imshow(dfx_lam(X, Y), extent=[xx[0], xx[-1], xx[0], xx[-1]], origin="upper")
axs[1].set_title(f"$f_x = {sym.latex(df_x)}$")

axs[2].imshow(dfy_lam(X, Y), extent=[xx[0], xx[-1], xx[0], xx[-1]], origin="upper")
axs[2].set_title(f"$f_y = {sym.latex(df_y)}$")

for a in axs:
    a.set_xlabel("x")
    a.set_ylabel("y")

plt.tight_layout()
plt.show()
```
