# 2D Functions.

Visualize the 2D Function (3 Variables) is 2-D space.

$$Z = sin(X) + cos(Y)$$

$$D_x,D_y = [0, 2\pi]$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = np.pi/6
xx = np.arange(0, 2 * np.pi + dx/3, dx)

Z = np.zeros((len(xx), len(xx)))

for coli, x in enumerate(xx):
  for rowi, y in enumerate(xx):
    # y, x not x, y
    Z[rowi, coli] = np.sin(x) + np.cos(y)

plt.imshow(Z, extent=(xx[0], xx[-1], xx[0], xx[-1]))
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Z = sin(x) + cos(y)')

plt.show()
```

### Don't Use Double For Loops

```python
dx = np.pi/6
xx = np.arange(0, 2 * np.pi + dx/3, dx)

X, Y = np.meshgrid(xx, xx)

# Create 1x2 subplot
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot X grid (x values constant in rows)
im1 = axs[0].imshow(X, extent=(0, 2*np.pi, 0, 2*np.pi), origin='lower')
axs[0].set_title('X Grid (x values vary across columns)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
fig.colorbar(im1, ax=axs[0])

# Plot Y grid (y values constant in columns)
im2 = axs[1].imshow(Y, extent=(0, 2*np.pi, 0, 2*np.pi), origin='lower')
axs[1].set_title('Y Grid (y values vary across rows)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
fig.colorbar(im2, ax=axs[1])

plt.tight_layout()
plt.show()
```

```python
Z = np.sin(X) + np.cos(Y)

plt.imshow(Z, extent=(xx[0], xx[-1], xx[0], xx[-1]))
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Z = sin(x) + cos(y)')

plt.show()
```
