# 2D Functions Sympy.

```python
import sympy as sym
from sympy.plotting import plot3d
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
```

$$f(x,y) = x^2 + \sin(y)$$

```python
x, y = sym.symbols('x y')
fxy = x**2 + sym.sin(y)
fxy
```

```python
plot3d(fxy, (x, -1.5, 1.5), (y, -sym.pi, sym.pi),
       title=f"$f(x,y)={sym.latex(fxy)}$")
plt.show()
```

$$Z = sin(X + Y^2)$$
$$D_x, D_y = [-\pi,\pi]$$

```python
fxy = sym.sin(x + y**2)

plot3d(fxy, (x, -sym.pi, sym.pi), (y, -sym.pi, sym.pi))
plt.show()
```

```python
f_num = sym.lambdify((x, y), fxy, modules='numpy')

xx = np.linspace(-np.pi, np.pi, 40)
X, Y = np.meshgrid(xx, xx)
Z = f_num(X, Y)


# Step 5: Create the Plotly surface plot
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z)])

fig.update_layout(
  title='3D Plot of sin(x + y)',
  scene=dict(
    xaxis_title='x',
    yaxis_title='y',
    zaxis_title='z = sin(x + y)'
  )
)

fig.show()
```
