# 2D Functions.

Visualize the 2D Function (3 Variables) is 2-D space.

$$Z = sin(X) + cos(Y)$$

$$D_x,D_y = [0, 2\pi]$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
dx = np.pi/6
xx = np.arange(0, 2 * np.pi + dx/3, dx)
X, Y = np.meshgrid(xx, xx)
Z = np.sin(X) + np.cos(Y)
```

```python
# Step 3: Create 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Step 4: Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Step 5: Add labels and colorbar
ax.set_title('z = sin(x) + cos(y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.show()
```
