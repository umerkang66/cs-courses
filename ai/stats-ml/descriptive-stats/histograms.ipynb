### Histograms

How many Bins should be in a Histogram Plot.

$$k = \lceil \frac{max(x) - min(x)}{h} \rceil$$

**Sturges**: Depends on the data Count.

$$k = \lceil log_2(n)\rceil + 1$$

**Freedman-Diaconis**: Depends on data count and on data spread. *Usually the best ONE*.

$$h = 2\frac{IQR}{\sqrt[3]{n}}$$

**Arbitrary**

$$k = 40$$

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
n = 1000
k = 40 # no of bins

data = np.exp(np.random.randn(n)/2)

plt.hist(data, k)
plt.xlabel('Value')
plt.ylabel('Count')
plt.show()
```

```python
# freeman diaconis rule
h = 2*stats.iqr(data)*n**(-1/3)
k = np.ceil( (np.max(data)-np.min(data))/h )

# plt.hist(data, bins='fd')

plt.hist(data, int(k))
plt.xlabel('Value')
plt.ylabel('Count')
plt.title(f"With k: {int(k)}")
plt.show()
```
