```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Gaussian Distribution

```python
n = 1001

x = np.linspace(-4, 4, n)
# convert into gaussian
gaussian_dist = stats.norm.pdf(x)

plt.plot(x, gaussian_dist)
plt.title("Analytic Gaussian (Normal) distribution")
plt.show()

print(gaussian_dist.sum())
print("This is more than 1, so it is not a probability distribution")
```

```python
# GAUSSIAN, in probability

n = 1001

x = np.linspace(-4, 4, n)
# convert into gaussian
gaussian_dist = stats.norm.pdf(x)
gaussian_dist_total = gaussian_dist.sum()
gaussian_dist = gaussian_dist / gaussian_dist_total

plt.plot(x, gaussian_dist)
plt.title("Analytic Gaussian (Normal) distribution")
plt.show()

print(gaussian_dist.sum())
print("This is equal to 1, so it is a probability distribution")
```

```python
# GAUSSIAN, normalizing using calculus
# Based on a Formula

n = 1001

x = np.linspace(-4, 4, n)
# convert into gaussian
gaussian_dist = stats.norm.pdf(x)

# AREA UNDER THE CURVE
print(np.sum(gaussian_dist * np.diff(x[:2])))

# We can also plot these values, these will be turned out as the gaussian distribution
```

```python
# Normally Distributed Random Numbers

stretch = 1 # variance (square of standard deviation)
shift = 5 # mean
n = 1000

# create data
data = stretch * np.random.randn(n) + shift
data2 =  np.random.randn(n)

plt.hist(data, bins=25)
plt.title("Empirical Normal Distribution")
plt.show()
```

```python
# Uniformly Distributed Numbers

stretch = 2
shift = .5
n = 10000

data = stretch * np.random.rand(n) + shift-stretch/2

fig,ax = plt.subplots(2, 1, figsize=(5, 6))

ax[0].plot(data, '.', markersize=1)
ax[0].set_title("Uniform Data Histogram")

ax[1].hist(data, bins=25, edgecolor='k')
ax[1].set_title("Uniform Data Histogram")

plt.show()
```

```python
# Log-Normal Distribution

n = 1001
x = np.linspace(0, 10, n)
log_normal_dist = stats.lognorm.pdf(x, 1)

plt.plot(x, log_normal_dist)
plt.title("Analytic Log Normal Distribution")
plt.show()
```

```python
# Log-Normal Distribution

n = 2000
shift = 5
stretch = .5

data = stretch * np.random.randn(n) + shift
data = np.exp(data)

fig,ax = plt.subplots(2, 1, figsize=(5, 6))

ax[0].plot(data, '.')
ax[0].set_title("Log Normal Data Values")

ax[1].hist(data, bins=30)
ax[1].set_title("Log Normal Data Histogram")

plt.show()
```

## Binomial Distribution

```python
# Binomial Distribution is the probability of K Heads in N coin tosses
# given the probability of heads

n = 10 # number of coin flips
p = .5 # probability of heads

x = np.arange(n + 2)
binomial_dist = stats.binom.pmf(x, n, p)

plt.bar(x, binomial_dist)
plt.title("Binomial Distribution (n=%s, p=%g)"%(n,p))

plt.show()
```

## T Distribution

```python
x = np.linspace(-4,4,1001)
df = 200
t = stats.t.pdf(x, df)

plt.plot(x, t)
plt.xlabel("t-value")
plt.ylabel("P(t | H$_0$)")
plt.title("t(%g) distribution"%df)

plt.show()
```

## F-Distribution

```python
num_df = 5 # numerator degrees of freedom
den_df = 100 # denominator df

x = np.linspace(0, 10, 10001)

f_dist = stats.f.pdf(x, num_df, den_df)

plt.plot(x, f_dist)
plt.title("F({num_df},{den_df}) distribution")
plt.xlabel("F value")
plt.show()
```
