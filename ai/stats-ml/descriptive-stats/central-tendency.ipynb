```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Create some data distributions

N = 10001
nbins = 30

d1 = np.random.randn(N) - 1
d2 = np.random.randn(N)
d3 = np.random.randn(N) + 1

y1,x1 = np.histogram(d1, bins=nbins)
x1 = (x1[1:] + x1[:-1]) / 2

y2,x2 = np.histogram(d2, bins=nbins)
x2 = (x2[1:] + x2[:-1]) / 2

y3,x3 = np.histogram(d3, bins=nbins)
x3 = (x3[1:] + x3[:-1]) / 2

plt.plot(x1, y1, 'r', label="-1")
plt.plot(x2, y2, 'g', label="0")
plt.plot(x3, y3, 'b', label="1")

plt.legend()
plt.xlabel("Data Values")
plt.ylabel("Data Counts")
plt.show()
```

```python
# also plotting the mean

plt.plot(x1, y1, 'r', label="-1")
plt.plot(x2, y2, 'g', label="0")
plt.plot(x3, y3, 'b', label="1")

plt.plot([d1.mean(), d1.mean()], [0, y1.max()], 'r--', label="Mean -1")
plt.plot([d2.mean(), d2.mean()], [0, y2.max()], 'g--', label="Mean 0")
plt.plot([d3.mean(), d3.mean()], [0, y3.max()], 'b--', label="Mean 1")


plt.legend()
plt.xlabel("Data Values")
plt.ylabel("Data Counts")
plt.show()
```

```python
d4 = np.hstack( (np.random.randn(N)-2, np.random.randn(N)+2) )

y4,x4 = np.histogram(d4, bins=nbins)
x4 = (x4[1:] + x4[:-1]) / 2

plt.plot(x4,y4, 'g', label="d4")
plt.plot([x4.mean(), x4.mean()], [0, y4.max()], 'g--', label="Mean d4")

plt.legend()
plt.xlabel("Data Values")
plt.ylabel("Data Counts")
plt.show()
```

```python
# median
shift = 0
stretch = .7
n = 2000
nbins = 50

# generate the data
data = stretch * np.random.randn(n) + shift
data = np.exp(data)

y,x = np.histogram(data)
x = (x[1:] + x[:-1]) / 2

datamean = np.mean(data)
datamedian = np.median(data)

fig,ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].plot(data, '.', color=[.5,.5,.5], label="data")
ax[0].plot([1, n], [datamean, datamean], 'b--', label="mean")
ax[0].plot([1, n], [datamedian, datamedian], 'g--', label="median")
ax[0].legend()

ax[1].plot(x, y, label="data")
ax[1].plot([datamean, datamean], [0, y.max()], 'b--', label="mean")
ax[1].plot([datamedian, datamedian], [0, y.max()], 'g--', label="median")
ax[1].legend()
plt.show()
```
