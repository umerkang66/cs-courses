```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
n = 1000
# log-normal distribution
data = np.exp(np.random.randn(n)/2)
```

```python
# just for analysis
plt.plot(np.arange(n), data, 's')

plt.title("Log Normal Distribution")
plt.xlabel("Range")
plt.ylabel("Value")
plt.show()
```

```python
k = 20
plt.hist(data, bins=k, width=.22)

plt.title("Log Normal Distribution")
plt.ylabel("Range")
plt.xlabel("Value")
plt.show()
```

```python
# creating multiple log-normal distributions
m = 100
n = 1000

for i in range(m):
  data = np.exp(np.random.randn(n)/2)
  y,x = np.histogram(data, bins=k)
  xx = (x[1:] + x[:-1]) / 2
  plt.plot(xx, y)

plt.xlabel("Range")
plt.ylabel("Value")
plt.title("%i Log Normal Distributions (Histograms)"%m)
plt.show()
```
