```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
m = 30
n = 10

data = np.zeros((30, 10))

for i in range(n):
  data[:,i] = 30 * np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
sns.boxplot(data, orient='v')
plt.title("Box Plots of Each Columns")

plt.show()
```

```python
df = pd.DataFrame(data, columns=[str(i) for i in np.arange(0, 10)])

sns.boxplot(data=df, orient='h')
plt.title("Box Plots of Each Columns")

plt.show()
```

### Challenge: Normal and Uniform Distribution Box Plot

```python
n = 100
data_gaussian = np.random.randn(n, 1)
data_normal = np.random.rand(n, 1)
```

```python
plt.boxplot(data_gaussian, positions=[1], widths=0.4, patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'))
plt.boxplot(data_normal, positions=[2], widths=0.4, patch_artist=True, boxprops=dict(facecolor='lightgreen', color='green'))

plt.xticks([1, 2], ["Gaussian", "Normal"])
plt.ylabel("Values")
plt.title("Box Plots of Gaussian and Normal Data")
plt.show()
```
