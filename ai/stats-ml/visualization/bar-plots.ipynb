```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
m = 30
n = 10

data = np.zeros((30, 10))

for i in range(n):
  data[:,i] = 30 * np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
fig,ax = plt.subplots(1, 3, figsize=(14, 4))

ax[0].bar(np.arange(n), np.mean(data, axis=0))
ax[0].set_title("Bar Plot")

ax[1].errorbar(np.arange(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[1].set_title("Error Plot")

ax[2].bar(np.arange(n), np.mean(data, axis=0))
ax[2].errorbar(np.arange(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='',color='k')
ax[2].set_title("Bar+Error Plot")

plt.show()
```

```python
xcrossings = np.array([1, 2, 4, 5, 6, 9])

# select only the xcrossing columns from the data
plt.bar(xcrossings, np.mean(data[:,xcrossings], axis=0))
plt.errorbar(xcrossings, np.mean(data[:,xcrossings], axis=0), np.std(data[:,xcrossings], axis=0, ddof=1), marker='s', linestyle='',color='k')
plt.title("Bar Plot with custom crossings")

plt.show()
```

```python
# bars from matrices
M = np.array([ [2, 5, 4, 3], [1, 1, 8, 6] ])

fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

ax[0, 0].imshow(M)
ax[0, 0].set_title("Showing as image")
# using pandas dataframe
df = pd.DataFrame(M, columns=["Col 0", "Col 1", "Col 2", "Col 3"])
df.plot(ax=ax[1,0], kind='bar')
ax[1, 0].set_title("Grouping by rows")


ax[0, 1].imshow(M.T)
ax[0, 1].set_title("T. Showing as image")
# using pandas dataframe
df = pd.DataFrame(M, columns=["Col 0", "Col 1", "Col 2", "Col 3"])
df.T.plot(ax=ax[1,1], kind='bar')
ax[1, 1].set_title("T. Grouping by rows")


plt.show()
```
