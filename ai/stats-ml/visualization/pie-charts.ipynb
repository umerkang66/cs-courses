```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nbins = 5
totalN = 100

rawdata = np.ceil(np.logspace(np.log10(1/2), np.log10(nbins-.01), totalN))

uniquenums = np.unique(rawdata)
data4pie = np.zeros(len(uniquenums))

for i in range(len(uniquenums)):
  data4pie[i] = np.sum(rawdata==uniquenums[i])

print(uniquenums)
print(data4pie)
```

```python
plt.pie(data4pie, labels=[str(i) + "%" for i in 100*data4pie/(np.sum(data4pie))])
plt.show()
```

```python
plt.pie(data4pie, labels=np.arange(5))
plt.show()
```
